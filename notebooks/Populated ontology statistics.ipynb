```python
from rdflib import Graph as RDFGraph
from rdflib.extras.external_graph_libs import rdflib_to_networkx_graph
import networkx as nx
from networkx import Graph as NXGraph
import matplotlib.pyplot as plt
import statistics
import collections
```

```python
print("Reading RDF file and parsing as an rdflib Graph")
rdf_path = "../comptox_aop.rdf"
rg = RDFGraph()
rg.parse(rdf_path, format='xml')
print("rdflib Graph loaded successfully with {0} triples".format(len(rg)))
```

```python
print("Converting rdflib Graph to NetworkX Graph")
G = rdflib_to_networkx_graph(rg)
print("NetworkX Graph loaded successfully with length {0}".format(len(G)))
```

```python
def mean(numbers):
    return float(sum(numbers)) / max(len(numbers), 1)

def num_leaf_nodes(g):
    """Returns the number of nodes with a degree of 1."""
    leaf_nodes = 0
    for u in g:
        if g.degree[u] == 1:
            leaf_nodes += 1
    return leaf_nodes

```

```python
# Network size
print("NETWORK SIZE")
print("============")
print("The network has {} nodes and {} edges".format(G.number_of_nodes(), G.number_of_edges()))
print()

# Network size
print("LEAF NODES")
print("============")
print("The network has {} leaf nodes".format(num_leaf_nodes(G)))
print()

# Density
print("DENSITY")
print("============")
print("The network density is {}".format(nx.density(G)))
print()

# Degree centrality -- mean and stdev
dc = nx.degree_centrality(G)
degrees = []
for k,v in dc.items():
    degrees.append(v)
    
print("DEGREE CENTRALITY")
print("=================")
print("The mean degree centrality is {}, with stdev {}".format(mean(degrees), statistics.stdev(degrees)))
print("The densest node is {}, with value {}".format(max(dc, key=dc.get), max(dc.values())))
print("The sparsest node is {}, with value {}".format(min(dc, key=dc.get), min(dc.values())))

```
